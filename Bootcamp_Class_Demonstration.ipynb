**Data cleaning**

```python
dirty= pd.read_csv('https://raw.githubusercontent.com/KhmAsif/IMSE-441-Spring-2026/refs/heads/main/Coffee_dirty.csv')
dirty.head(3)
```

```python
dirty.info()
```

**Null values**

```python
dirty.isnull().sum()
```

```python
dirty=dirty.dropna()
dirty.isnull().sum()
```

**Duplicates**

```python
dirty.duplicated().sum()
```

```python
dirty=dirty.drop_duplicates()
dirty.duplicated().sum()
```

```python
dirty.info()
```

```python
dirty.columns = (dirty.columns
                 .str.strip()
                 .str.replace(" ", "_", regex=False)
                 .str.replace("__+", "_", regex=True)
                 .str.lower())

# coerce types
num_cols = ["inventory","sales","marketing","profit","unit_price","revenue"]
for c in num_cols:
    if c in dirty.columns:
        dirty[c] = (dirty[c]
            .astype(str)
            .str.replace(r"[$,]", "", regex=True)  # remove $ and commas
            .replace("", np.nan)
            .astype(float))
```

**Outlier removal**

```python
#define outlier removal function
def remove_outliers_iqr(df, column, k=1.5):
    q1 = Coffee[column].quantile(0.25)
    q3 = Coffee[column].quantile(0.75)
    iqr = q3 - q1

    lower = q1 - k * iqr
    upper = q3 + k * iqr

    return Coffee[(Coffee[column].between(lower, upper))]
```

```python
for col in ["marketing", "profit"]:
    Coffee = remove_outliers_iqr(Coffee, col)
```

```python
plt.boxplot([Coffee['marketing'],Coffee['profit']], tick_labels=['marketing', 'profit'])
plt.xlabel('Items')
plt.ylabel('Dollars')
plt.title('Boxplot')
plt.show()
```
